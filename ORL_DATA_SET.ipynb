```python
import os
import cv2
import numpy as np
import csv

# Path to the root directory containing the folders s1, s2, ..., s40
dataset_path = '/kaggle/input/att-database-of-faces'

# Path to save the CSV file
csv_file_path = 'image_pixel_data.csv'

# List to store the pixel data with labels
csv_data = []

# Iterate through each folder
for folder in os.listdir(dataset_path):
    folder_path = os.path.join(dataset_path, folder)
    
    # Check if the folder path is a directory
    if os.path.isdir(folder_path):
        label = folder  # Use the folder name (e.g., 's1') as the label
        
        # Iterate through each image in the folder
        for image_name in os.listdir(folder_path):
            image_path = os.path.join(folder_path, image_name)
            
            # Read the image in grayscale
            img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            
            # Check if the image is successfully loaded
            if img is not None:
                # Flatten the image pixels and prepend the label
                pixel_values = img.flatten().tolist()
                csv_data.append([label] + pixel_values)

# Save the data into a CSV file
with open(csv_file_path, mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    
    # Write the header row (Label + Pixel Columns)
    header = ['Label'] + [f'Pixel_{i}' for i in range(len(csv_data[0]) - 1)]
    csv_writer.writerow(header)
    
    # Write the pixel data with labels
    csv_writer.writerows(csv_data)

print(f"Data successfully saved to {csv_file_path}")
```
